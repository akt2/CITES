```python
import pandas as pd
```

```python
url = 'https://awionline.org/content/list-endangered-species'
```

```python
tables = pd.read_html(url)
tables
```

```python
type(tables)
```

```python
mammals_df = tables[0]
mammals_df.columns = ['Common Name', 'Scientific Name', 'Status']
mammals_df.head()
```

```python
birds_df = tables[1]
birds_df.columns = ['Common Name', 'Scientific Name','Status']
birds_df.head()
```

```python
reptiles_df = tables[2]
reptiles_df.columns = ['Common Name', 'Scientific Name', 'Status']
reptiles_df.head()
```

```python
amphibians_df = tables[3]
amphibians_df.columns = ['Common Name', 'Scientific Name', 'Status']
amphibians_df.head()
```

```python
fish_df = tables[4]
fish_df.columns = ['Common Name', 'Scientific Name', 'Status']
fish_df.head()
```

```python
clams_df = tables[5]
clams_df.columns = ['Common Name', 'Scientific Name', 'Status']
clams_df.head()
```

```python
snails_df = tables[6]
snails_df.columns = ['Common Name', 'Scientific Name', 'Status']
snails_df.head()
```

```python
insects_df = tables[7]
insects_df.columns = ['Common Name', 'Scientific Name', 'Status']
insects_df.head()
```

```python
arachnids_df = tables[8]
arachnids_df.columns = ['Common Name', 'Scientific Name', 'Status']
arachnids_df.head()
```

```python
crustaceans_df = tables[9]
crustaceans_df.columns = ['Common Name', 'Scientific Name', 'Status']
crustaceans_df.head()
```

```python
corals_df = tables[10]
corals_df.columns = ['Common Name', 'Scientific Name', 'Status']
corals_df.head()
```

```python
endangered_df = pd.concat([mammals_df, birds_df, reptiles_df, amphibians_df, fish_df, clams_df, snails_df, insects_df, arachnids_df, crustaceans_df, corals_df], ignore_index=True)
```

```python
endangered_df.set_index('Scientific Name', inplace=True)
```

```python
endangered_df.head()
```

```python
endangered_df.to_csv('animalwelfare.csv')
```
